# Reshaping Numpy arrays

<br> </br>

#### .reshape()

```python
a4 = np.arange(1,7)
a4
```

```python
a4.reshape(3,2)
```

```python
a4.reshape(2,3)
```

```python
a4.reshape(3,2, order = 'F')
```

#### .ravel()

```python
a5 = np.arange(1,7).reshape(3,2)
a5
```

```python
a5.ravel()
```

```python
a5.ravel(order = 'F')
```

#### .flatten()

```python
a5
```

```python
a5.flatten()
```

```python
a5.flatten(order = 'F')
```
