<h2>A hospital laboratory is trying to determine how many patients underwent three different medical tests in a day:

Blood Tests, X-Rays, MRI Scans

At the end of the day, they only know the aggregate operational data:

Total patients tested: 180
Total technician time consumed: 390 hours
Blood Test takes 1 hour

MRI takes 4 hours
Total billing amount collected: ₹1,530,000

X-Ray costs ₹5,000

```python
# x + y + z = 180
# x + 2y + 4z = 390
# 2000x + 5000y + 12000z = 1530000
```

```python
coeff = np.array([
    [ 1, 1, 1],
    [1, 2, 4],
    [2000, 5000, 12000]
])

values = np.array([180, 390, 1530000]).reshape(3,1)
```

```python
coeff
```

```python
values
```

## Smoothing Apple stock price trend

```python
import yfinance as yf
```

```python
# Download last 5 years stock data
stock_name = 'AAPL'
data = yf.download(
    stock_name,
    period="5y",
    interval="1d"
)

# Convert closing prices to NumPy array
prices = data["Close"].to_numpy()
```

```python
prices = prices.reshape(-1)
prices
```

```python
plt.plot(prices)
```

```python
x = np.linspace(-3, 3, 200)
sigma = 1
kernel = np.exp(-(x/sigma)**2)
kernel = kernel / np.sum(kernel)
```

```python
plt.plot(kernel)
```

```python
result = np.convolve(np.pad(prices, 100,mode = 'edge'), kernel, mode = 'valid')
plt.plot(result)
```

```python
plt.plot(prices, alpha = 0.4)
plt.plot(result, '--r')
```

```python
sigma = 2
kernel = np.exp(-(x/sigma)**2)
kernel = kernel / np.sum(kernel)

result = np.convolve(np.pad(prices, 100,mode = 'edge'), kernel, mode = 'valid')

plt.plot(prices, alpha = 0.4)
plt.plot(result, '--r')
```
